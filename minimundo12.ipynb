# Minimundo 12 — Qualidade da Água em Rios

Este notebook realiza:
- Classificação dos rios por nível de poluição
- Ranking dos rios
- Gráfico de poluição

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ambiental_rios.csv')

def classificar(ind):
    if ind <= 45:
        return 'Seguro'
    elif 46 <= ind <= 75:
        return 'Alerta'
    else:
        return 'Crítico'

df['Classificação'] = df['indice_poluição'].apply(classificar)

df_ranking = df.sort_values(by='indice_poluição', ascending=False)
df_ranking
```

```python
plt.figure(figsize=(12,6))
plt.bar(df_ranking['rio'], df_ranking['indice_poluição'])
plt.xticks(rotation=45, ha='right')
plt.title('Ranking de Poluição dos Rios')
plt.xlabel('Rio')
plt.ylabel('Índice de Poluição')
plt.tight_layout()
plt.show()
```
